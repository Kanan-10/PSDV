Q2. Immutable Datatypes: These are data types whose values cannot be changed once the object is created. Any modification to an immutable object results in the creation of a new object with a new identity. For example, strings, tuples, int, float.

Mutable Datatypes: These are data types whose values can be changed or modified after the object is created. When a mutable object is altered, it retains its identity. For example, lists, sets, dictionaries.

```python
#Q20
import csv
import random
import matplotlib.pyplot as plt

# Generate 20 random integers between 1 and 20
random_integers = [random.randint(1, 20) for _ in range(20)]

# Write the integers to a CSV file
with open('random_integers.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Index', 'Value'])  # Writing the header
    for index, value in enumerate(random_integers):
        writer.writerow([index, value])

# Read the integers from the CSV file and compute the sum of the 'Value' column
sum_values = 0
read_values = []

with open('random_integers.csv', mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        value = int(row['Value'])
        sum_values += value
        read_values.append(value)

print(f"Sum of values in 'Value' column: {sum_values}")

# Visualize the data using a line plot
plt.plot(read_values, marker='o')
plt.show()
```
